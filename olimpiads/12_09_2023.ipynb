```python
# грядки
l = 7           # расстояние от колодца до первой грядки
n = 5           # длина грядки
m = 10          # ширина грядки
P = 2 * (m + n) # периметр грядки
k = 20          # количество грядок
S = 0           # общее расстояние обхода
```

```python
# решение при помощи цикла
for _ in range(k):
  S += 2 * l + (_ * 2 * m) + P
print(S)
```

```python
# решение формулой
S = k * 2 * l + k * P + (k - 1) * k * m
print(S)
```
